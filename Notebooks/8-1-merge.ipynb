```python
pluto = pd.read_csv("../Data/Source Data/pluto_25v1.csv", low_memory=False)
```

Before merging datasets, it's a good idea to check that the keys in each dataset match those in the other:

```python
(
    work_on_stop_work_order['bbl']
    .isin(pluto['bbl'])
    .all()
)
```

```python
(
    work_on_stop_work_order
    [
        ~
        work_on_stop_work_order['bbl'].isin(pluto['bbl'])
    ]
    ['bbl']
)
```

```python
(
    work_on_stop_work_order['bbl'].dropna()
    .isin(pluto['bbl'])
    .all()
)
```

If keys do not exactly match, use string manipulation tools (like `.str.replace()`, `.str.upper()`, `.str.split()` etc. to _normalize_ the labels between the two. This is an art as much as a science.)

Check for overlapping columns (by finding `intersection` of two sets of columns names)

```python
set(work_on_stop_work_order.columns).intersection(set(pluto.columns))
```

Column names that overlap and are not used for the merge are, by default, both kept but given the distinguishing suffixes `_x` and `_y`. 
Specifying `suffixes=` will apply custom suffixes to (only) the overlapping column names.

Use `validate` to ensure that keys are unique on one or both source tables. `1:1` checks that keys are unique on both sides (for an exact *one-to-one* merge). `m:1` checks that keys are unique on the right; these rows are broadcast to each matching row on the left: a *many-to-one* merge (`1:m` checks the opposite, for a one-to-many merge).

Using this validation argument avoid accidentally merging on non-unique keys.

```python
work_on_stop_pluto = (
    work_on_stop_work_order
    .merge(
        pluto,
        on='bbl',
        how='inner',
        validate='m:1',
        suffixes=[None,'_pluto']
    )
)
```

Once you have a merged DataFrame you can use combine columns from both sources for analysis:

```python
(
    work_on_stop_pluto
    .groupby('borough')
    ['unique_key']
    .nunique()
)
```

```python
(
    work_on_stop_pluto
    .groupby('zonedist1')
    ['unique_key']
    .nunique()
    .sort_values(ascending=False)
    .head(5)
)
```

If the matching key(s) have different column labels in the left and right tables, the key columns for each can be specified with `left_on` and `right_on` (or, if either has the matching key in an index, that can be specified with `left_index=True` or `right_index=True`)

Note:
- Merges will only link _exact_ matches between keys:
  - keys in `int` or `float` numeric types _are_ equivalent and will be matched.
  - but keys in numeric types will not be matched to `str` types. Cast one or the other column to a matching data type.
  - `str` keys will only match if they are _exact_ matches. Check for differences in capitalization, punctuation, etc. before merging on text keys (especially for data from different sources or that has been manually-entered).
  - datetime keys will match, but only to keys at the same detail (e.g. dates with dates, timestamps with timestamps). Resample datatime data to a common precision before merging. 

the `.join()` method is an abbreviated form of the `.merge()` method which joins DataFrames on their indexes (only)

## concat

`concat` (concatenate) extends datasets longer by appending datasets together

E.g. litigation regarding Department of Corrections, published twice-annually by Law Department: https://www.nyc.gov/site/law/public-resources/nyc-administrative-code-7-113.page

```python
doc_litigation_fy_20_24 = pd.read_excel('https://www.nyc.gov/assets/law/downloads/excel/DOC%20Alleged%20Misconduct%20Matters%20Commenced%20in%20FY%202020-2024.xls')
doc_litigation_cy_20_24 = pd.read_excel('https://www.nyc.gov/assets/law/downloads/excel/DOC%20Alleged%20Misconduct%20Matters%20Commenced%20in%20CY%202020-2024.xlsx')
```

```python
doc_litigation_cy_20_24
```

`pd.concat()` will align DataFrames on columns. If the DataFrames to combine have matching data fields in columns with different names (e.g. if you want to a column `Date` in the first DataFrame to be matched with a columns `month` in the second), use `.rename()` first to make hte column names match.

```python
doc_litigation_fy_20_24.columns.equals(doc_litigation_cy_20_24.columns)
```

```python
doc_litigation = pd.concat([
    doc_litigation_fy_20_24,
    doc_litigation_cy_20_24
])
```

```python
doc_litigation
```

```python
doc_litigation = (
    doc_litigation
    .drop_duplicates(
        keep='last'
    )
)
```

If data rows are overlapping across the concatenated sources, use `.drop_duplicates()` to remove duplicates. (include a `subset=` to drop only rows duplicated on that column)
